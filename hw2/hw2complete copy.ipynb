```python
import numpy as np 
#from numpy import array
from scipy.linalg import solve
from scipy.linalg import lu

#PROBLEM 1

m = 5 ; n = 3


A = np.random.randint(-2,5 , size=(m, n) )

b = np.random.randint(-2,3, size=(m, 1))


ATA = A.T @ A


AAT = A @ A.T

ATb = A.T @ b



P, L, U = lu(ATA)

P1, L1, U1 = lu(AAT)


#solve ATAx = ATb
y = solve(L, P.T @ ATb)
x = solve(U, y)

#solve AATx = b, AAT  - problem is singular
y1 = solve(L1, P1.T @ b)
x1 = solve(U1, y1)

print("ATA solution")
print(x)
print(" ")
print("AAT solution")
print(x1)
```

```python
import numpy as np
from numpy.linalg import qr 
A1 = np.random.randint(-2,5 , size=(m, n))

A1 = A1.T @ A1



print("A1 = ")
print(A1)

MaxIter = 60; p = [1, 10, 20, MaxIter]
A = A1
for i in range(MaxIter):
    Q, R = qr(A)
    A = R @ Q

    if i+1 in p:
        print(f'Iteration {i+1}:')
        print(R)
        
print("A = ")
print(A)
```
