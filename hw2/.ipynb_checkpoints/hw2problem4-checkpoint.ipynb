```python
# EigenFactorization of a Matrix
from numpy import array
import numpy as np
from numpy.linalg import eig

m = 5 ; n = 3

A = np.random.randint(-2,5 , size=(m, n) )

 

print("Original Matrix")
print(A)
print(" ------------- ")

ATA = A.T @ A

Eig_Values, Eig_Vectors = eig(ATA)
#
print("EigenValues of ATA")
print(Eig_Values)
print(" ------------- ")
#
print("EigenVectors of ATA")
print(Eig_Vectors)
print(" ------------- ")
 
AAT = A @ A.T

Eig_Values, Eig_Vectors = eig(AAT)
#
print("EigenValues of AAT")
print(Eig_Values)
print(" ------------- ")
#
print("EigenVectors of AAT")
print(Eig_Vectors)
print(" ------------- ")
```

```python
import numpy as np
from numpy.linalg import qr 
#
A1 = np.random.randint(-2,5 , size=(m, n))

A1 = A1.T @ A1



print("A1 = ")
print(A1)

MaxIter = 60; p = [1, 10, 20, MaxIter]
A = A1
for i in range(MaxIter):
    Q, R = qr(A)
    A = R @ Q

    if i+1 in p:
        print(f'Iteration {i+1}:')
        print(R)

print("A = ")
print(A)
```
